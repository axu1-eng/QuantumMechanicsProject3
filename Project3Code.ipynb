#Setup (Defining Variables, Importing Libraries)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import cmath

p0 = 1
x0 = 0
numberOfSteps = 2001
xMin = -5
xMax = 5
stepSize = (xMax-xMin)/numberOfSteps
A = 1
```

#Wavefunction Initialisation

```python
def initialWaveFunction(x):
  return A*np.exp(p0*(x-x0)*1j)
initialisedWaveFunction = []
currentXValue = xMin
for i in range (numberOfSteps):
  initialisedWaveFunction.append(initialWaveFunction(currentXValue))
  currentXValue += stepSize
print(initialisedWaveFunction)
```
